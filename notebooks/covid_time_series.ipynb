```python
import pandas as pd
import numpy as np
```

```python
url_confirmed = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
url_deaths = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
url_recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"

# Load the data into DataFrames
df_confirmed = pd.read_csv(url_confirmed)
df_deaths = pd.read_csv(url_deaths)
df_recovered = pd.read_csv(url_recovered)

# Display the first few rows of the datasets
df_confirmed.head()
```

```python
# Drop unnecessary columns and group by country
df_confirmed = df_confirmed.drop(columns=['Province/State', 'Lat', 'Long'])
df_confirmed = df_confirmed.groupby('Country/Region').sum()

df_confirmed.head()
```

```python
# Transpose the DataFrame to have dates as rows and countries as columns
df_confirmed = df_confirmed.transpose()
```

```python
df_confirmed.tail()
```

```python
df = df_confirmed.copy()
df.reset_index(inplace=True)
df = df.rename(columns={'index': 'Date'})
df.head()
```

```python
# Convert the index to datetime
df_confirmed.index = pd.to_datetime(df_confirmed.index, format='%m/%d/%y')

# # Sort the DataFrame by date
# df_confirmed = df_confirmed.sort_index(

# Display the first few rows of the transformed dataset
df_confirmed.tail()
```

```python
import matplotlib.pyplot as plt

# Function to plot time series data for a given country using subplots
def plot_time_series_subplots(countries):
    num_countries = len(countries)
    fig, axes = plt.subplots(num_countries, 1, figsize=(10, 6 * num_countries), sharex=True)

    for i, country in enumerate(countries):
        axes[i].plot(df.index, df[country], label=country)
        axes[i].set_xlabel('Date')
        axes[i].set_ylabel('Confirmed Cases')
        axes[i].set_title(f'COVID-19 Confirmed Cases in {country}')
        axes[i].legend()

    plt.tight_layout()
    plt.show()

# Plot time series data for a few countries
plot_time_series_subplots(['US', 'India', 'Brazil'])
```

```python
def plot_multiple_countries(countries):
    plt.figure(figsize=(10, 6))
    for country in countries:
        plt.plot(df.index, df[country], label=country)
    plt.xlabel('Date')
    plt.ylabel('Confirmed Cases')
    plt.title('COVID-19 Confirmed Cases Comparison')
    plt.legend()
    plt.show()

# Plot time series data for multiple countries
plot_multiple_countries(['US', 'India', 'Brazil', 'Russia', 'South Africa'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def decompose_time_series(country):
    series = df[country]
    decomposition = seasonal_decompose(series, model='additive')

    # Plot decomposition
    plt.figure(figsize=(14, 10))
    decomposition.plot()
    plt.show()

# Decompose time series for the US
decompose_time_series('US')
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Function to fit and forecast using ARIMA
def arima_forecast(country, order=(5, 1, 0)):
    series = df[country]

    # Fit the ARIMA model
    model = ARIMA(series, order=order)
    model_fit = model.fit()

    # Forecast the next 30 days
    forecast = model_fit.forecast(steps=30)

    # Plot the forecast
    plt.figure(figsize=(10, 6))
    plt.plot(series, label='Actual')
    plt.plot(forecast, label='Forecast', color='red')
    plt.xlabel('Date')
    plt.ylabel('Confirmed Cases')
    plt.title(f'COVID-19 Forecast for {country}')
    plt.legend()
    plt.show()

# Forecast for the US
arima_forecast('US')
```

```python
from sklearn.metrics import mean_squared_error

# Function to evaluate ARIMA model
def evaluate_arima_model(country, order=(5, 1, 0), train_ratio=0.8):
    series = df[country]
    train_size = int(len(series) * train_ratio)
    train, test = series[:train_size], series[train_size:]

    # Fit the ARIMA model on the training set
    model = ARIMA(train, order=order)
    model_fit = model.fit()

    # Forecast on the test set
    forecast = model_fit.forecast(steps=len(test))

    # Calculate mean squared error
    mse = mean_squared_error(test, forecast)

    # Plot the forecast vs actual values
    plt.figure(figsize=(10, 6))
    plt.plot(train, label='Training Data')
    plt.plot(test, label='Test Data')
    plt.plot(test.index, forecast, label='Forecast', color='red')
    plt.xlabel('Date')
    plt.ylabel('Confirmed Cases')
    plt.title(f'COVID-19 Forecast Evaluation for {country}')
    plt.legend()
    plt.show()

    print(f'Mean Squared Error: {mse}')

# Evaluate ARIMA model for the US
evaluate_arima_model('US')
```
